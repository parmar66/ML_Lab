## Program to perform all dictionary Operations

```python
#Creating a dictionary
student = {
    "Name":"Harsh Parmar",
    "RollNo":3466,
    "Subject":"Machine Learning",
    "Marks":78
}
```

```python
#display Dictionary
print(student)
```

```python
#Access values using keys
print(("Name"),student["Name"])
print(("RollNo :"),student["RollNo"])
print(("Subject:"),student["Subject"])
```

```python
#update value 
student["Marks"]=85
print("Updated marks :",student["Marks"])
```

```python
#add new key value pair :
student["Grade "]="A"
print(student)
```

```python
#display all keys
print("keys :",student.keys())
```

```python
#display all values 
print("Values",student.values())
```

```python
#display key-value pairs
print("Items:",student.items())
```

```python
#length of dictionary
print("total keys:" ,len(student))
```

```python
#put data into dictionary
names = ["Riya","kiran","Siya","palak","Ajay"]
rollno=[1,2,3,4,5]
subject=["Maths","Science","English","C++","java"]
marks=[54,65,85,79,89]


```

```python
data = {}

for i in range(len(names)):
    data[rollno[i]] = {
        "name": names[i],
        "subject": subject[i],
        "marks": marks[i]
    }
```

```python
for rollno, info in data.items():
    print("Roll No:", rollno)
    print("Name:", info["name"])
    print("Subject:", info["subject"])
    print("Marks:", info["marks"])
    print("-" * 25)

    
```
